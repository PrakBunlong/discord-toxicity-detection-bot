```python
import pandas as pd
```

```python
df = pd.read_csv("../../dataset/unprocessed/jigsaw-toxic-comment-classification-challenge/train.csv")
df.head(20)
```

```python
def label_is_toxic(row):
    if row["toxic"] == 1 or row["severe_toxic"] == 1 or row["obscene"] == 1 or row["threat"] == 1 or row["insult"] == 1 or row["identity_hate"] == 1:
        return 1
    return 0
```

```python
df["toxic"]
```

```python
df["is_toxic"] = df.apply(label_is_toxic, axis=1)
```

```python
df = df.loc[:, ["id", "comment_text", "is_toxic"]]
df.head(20)
```

```python
df.to_csv("../../dataset/processed/jigsaw-toxic-comment-classification-challenge/train.csv")
```
